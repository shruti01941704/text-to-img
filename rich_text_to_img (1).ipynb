```python
# from PIL import ImageEnhance, ImageOps
# import torch
# from diffusers import StableDiffusionPipeline
# import io
# import base64
# import IPython.display as display
# import ipywidgets as widgets
# from IPython.display import HTML

# # Load the Stable Diffusion model
# pipeline = StableDiffusionPipeline.from_pretrained(
#     "stabilityai/stable-diffusion-2-1", torch_dtype=torch.float16
# ).to("cuda")

# # Function to generate image
# def generate_image(prompt):
#     with torch.no_grad():
#         image = pipeline(prompt=prompt).images[0]
#     return image

# # Function to apply brightness, contrast, saturation, and filters
# def modify_image(image, brightness_factor, contrast_factor, saturation_factor, selected_filter):
#     # Enhance brightness
#     enhancer = ImageEnhance.Brightness(image)
#     image = enhancer.enhance(brightness_factor)

#     # Enhance contrast
#     enhancer = ImageEnhance.Contrast(image)
#     image = enhancer.enhance(contrast_factor)

#     # Enhance saturation
#     enhancer = ImageEnhance.Color(image)
#     image = enhancer.enhance(saturation_factor)

#     # Apply additional filters
#     if selected_filter == 'Grayscale':
#         image = ImageOps.grayscale(image)
#     elif selected_filter == 'Sepia':
#         sepia = ImageOps.colorize(ImageOps.grayscale(image), '#704214', '#C0C090')
#         image = sepia
#     elif selected_filter == 'Invert':
#         image = ImageOps.invert(image.convert('RGB'))

#     return image

# # Function to convert image to base64 for HTML display
# def convert_to_base64(image):
#     img_byte_arr = io.BytesIO()
#     image.save(img_byte_arr, format='PNG')
#     img_byte_arr = img_byte_arr.getvalue()
#     return base64.b64encode(img_byte_arr).decode('utf-8')

# # Function to display image in the output
# def display_image(image):
#     img_base64 = convert_to_base64(image)
#     img_html = f'''
#     <div style="display: flex; justify-content: center; align-items: center; margin-top: 20px;">
#         <img src="data:image/png;base64,{img_base64}" style="border: 5px solid #000000; border-radius: 10px; padding: 10px; max-width: 100%; height: auto;">
#     </div>
#     '''
#     image_box.clear_output(wait=True)
#     with image_box:
#         display.display(HTML(img_html))

# # Function to create download link for image
# def create_download_button(img_base64, button_text, filename):
#     return f'''
#     <div style="display: flex; justify-content: center; margin-top: 10px;">
#         <a href="data:image/png;base64,{img_base64}" download="{filename}" style="text-decoration: none; padding: 10px 20px; background-color: #007BFF; color: white; border-radius: 5px; font-weight: bold; display: inline-flex; align-items: center; box-shadow: 0px 4px 6px rgba(0, 0, 0, 0.1); transition: background-color 0.3s;">
#             <span style="margin-right: 5px;">&#x1F4BE;</span> {button_text}
#         </a>
#     </div>
#     '''

# # Function to handle button click for generating image
# def on_generate_button_click(b):
#     prompt = input_box.value
#     if prompt:
#         global generated_image
#         generated_image = generate_image(prompt)

#         display_image(generated_image)

#         img_base64 = convert_to_base64(generated_image)
#         download_link = create_download_button(img_base64, "Download Original Image", "original_image.png")

#         download_box.clear_output()
#         with download_box:
#             display.display(HTML(download_link))

#         enable_modification_controls()

# # Function to handle button click for modifying image
# def on_modify_button_click(b):
#     brightness_factor = brightness_slider.value
#     contrast_factor = contrast_slider.value
#     saturation_factor = saturation_slider.value
#     selected_filter = filter_dropdown.value

#     modified_image = generated_image.copy()

#     # Apply enhancements and filters
#     modified_image = modify_image(modified_image, brightness_factor, contrast_factor, saturation_factor, selected_filter)

#     display_image(modified_image)

#     img_base64 = convert_to_base64(modified_image)
#     download_link = create_download_button(img_base64, "Download Modified Image", "modified_image.png")

#     download_box.clear_output()
#     with download_box:
#         display.display(HTML(download_link))

# # Function to enable modification controls
# def enable_modification_controls():
#     brightness_slider.disabled = False
#     contrast_slider.disabled = False
#     saturation_slider.disabled = False
#     filter_dropdown.disabled = False
#     modify_button.disabled = False

# # UI Elements for text prompt
# input_box = widgets.Text(
#     description='Prompt:',
#     placeholder='Enter your text here',
#     layout=widgets.Layout(width='50%')
# )

# # UI Elements for image enhancements
# brightness_slider = widgets.FloatSlider(
#     value=1.0,
#     min=0.0,
#     max=2.0,
#     step=0.1,
#     description='Brightness',
#     disabled=True
# )

# contrast_slider = widgets.FloatSlider(
#     value=1.0,
#     min=0.0,
#     max=2.0,
#     step=0.1,
#     description='Contrast',
#     disabled=True
# )

# saturation_slider = widgets.FloatSlider(
#     value=1.0,
#     min=0.0,
#     max=2.0,
#     step=0.1,
#     description='Saturation',
#     disabled=True
# )

# # Dropdown for filters
# filter_dropdown = widgets.Dropdown(
#     options=['None', 'Grayscale', 'Sepia', 'Invert'],
#     description='Filter',
#     disabled=True
# )

# # Buttons for generating and modifying the image
# generate_button = widgets.Button(description="Generate Image", button_style='primary')
# modify_button = widgets.Button(description="Modify Image", button_style='info')
# modify_button.disabled = True

# image_box = widgets.Output()
# download_box = widgets.Output()

# # Bind button click events to functions
# generate_button.on_click(on_generate_button_click)
# modify_button.on_click(on_modify_button_click)

# # Layout the UI
# ui = widgets.VBox([
#     widgets.HTML('<h2>Rich Text-to-Image Generator</h2>'),
#     input_box,
#     generate_button,
#     widgets.HTML('<h3>Modify Image</h3>'),
#     widgets.HBox([brightness_slider, contrast_slider, saturation_slider, filter_dropdown]),
#     modify_button,
#     image_box,
#     download_box
# ])

# # Display the UI
# display.display(ui)
from PIL import ImageEnhance, ImageOps
import torch
from diffusers import StableDiffusionPipeline
import io
import base64
import IPython.display as display
import ipywidgets as widgets
from IPython.display import HTML
import re

# Load the Stable Diffusion model
pipeline = StableDiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-2-1", torch_dtype=torch.float16
).to("cuda")

# Function to extract words with styles from HTML-like input
def parse_rich_text(prompt):
    # This regular expression will capture words with inline color styling
    styled_prompt = re.sub(r"<span style='color:(#[a-fA-F0-9]{6}|[a-zA-Z]+);'>(.*?)</span>",
                           lambda match: f'{match.group(2)} in {match.group(1)} color',
                           prompt)
    return styled_prompt

# Function to generate image
def generate_image(prompt):
    prompt = parse_rich_text(prompt)  # Parse rich text
    with torch.no_grad():
        image = pipeline(prompt=prompt).images[0]
    return image

# Function to apply brightness, contrast, saturation, and filters
def modify_image(image, brightness_factor, contrast_factor, saturation_factor, selected_filter):
    # Enhance brightness
    enhancer = ImageEnhance.Brightness(image)
    image = enhancer.enhance(brightness_factor)

    # Enhance contrast
    enhancer = ImageEnhance.Contrast(image)
    image = enhancer.enhance(contrast_factor)

    # Enhance saturation
    enhancer = ImageEnhance.Color(image)
    image = enhancer.enhance(saturation_factor)

    # Apply additional filters
    if selected_filter == 'Grayscale':
        image = ImageOps.grayscale(image)
    elif selected_filter == 'Sepia':
        sepia = ImageOps.colorize(ImageOps.grayscale(image), '#704214', '#C0C090')
        image = sepia
    elif selected_filter == 'Invert':
        image = ImageOps.invert(image.convert('RGB'))

    return image

# Function to convert image to base64 for HTML display
def convert_to_base64(image):
    img_byte_arr = io.BytesIO()
    image.save(img_byte_arr, format='PNG')
    img_byte_arr = img_byte_arr.getvalue()
    return base64.b64encode(img_byte_arr).decode('utf-8')

# Function to display image in the output
def display_image(image):
    img_base64 = convert_to_base64(image)
    img_html = f'''
    <div style="display: flex; justify-content: center; align-items: center; margin-top: 20px;">
        <img src="data:image/png;base64,{img_base64}" style="border: 5px solid #000000; border-radius: 10px; padding: 10px; max-width: 100%; height: auto;">
    </div>
    '''
    image_box.clear_output(wait=True)
    with image_box:
        display.display(HTML(img_html))

# Function to create download link for image
def create_download_button(img_base64, button_text, filename):
    return f'''
    <div style="display: flex; justify-content: center; margin-top: 10px;">
        <a href="data:image/png;base64,{img_base64}" download="{filename}" style="text-decoration: none; padding: 10px 20px; background-color: #007BFF; color: white; border-radius: 5px; font-weight: bold; display: inline-flex; align-items: center; box-shadow: 0px 4px 6px rgba(0, 0, 0, 0.1); transition: background-color 0.3s;">
            <span style="margin-right: 5px;">&#x1F4BE;</span> {button_text}
        </a>
    </div>
    '''

# Function to handle button click for generating image
def on_generate_button_click(b):
    prompt = input_box.value
    if prompt:
        global generated_image
        generated_image = generate_image(prompt)

        display_image(generated_image)

        img_base64 = convert_to_base64(generated_image)
        download_link = create_download_button(img_base64, "Download Original Image", "original_image.png")

        download_box.clear_output()
        with download_box:
            display.display(HTML(download_link))

        enable_modification_controls()

# Function to handle button click for modifying image
def on_modify_button_click(b):
    brightness_factor = brightness_slider.value
    contrast_factor = contrast_slider.value
    saturation_factor = saturation_slider.value
    selected_filter = filter_dropdown.value

    modified_image = generated_image.copy()

    # Apply enhancements and filters
    modified_image = modify_image(modified_image, brightness_factor, contrast_factor, saturation_factor, selected_filter)

    display_image(modified_image)

    img_base64 = convert_to_base64(modified_image)
    download_link = create_download_button(img_base64, "Download Modified Image", "modified_image.png")

    download_box.clear_output()
    with download_box:
        display.display(HTML(download_link))

# Function to enable modification controls
def enable_modification_controls():
    brightness_slider.disabled = False
    contrast_slider.disabled = False
    saturation_slider.disabled = False
    filter_dropdown.disabled = False
    modify_button.disabled = False

# UI Elements for text prompt
input_box = widgets.Text(
    description='Prompt:',
    placeholder="Enter text like: <span style='color:red;'>bird</span>",
    layout=widgets.Layout(width='50%')
)

# UI Elements for image enhancements
brightness_slider = widgets.FloatSlider(
    value=1.0,
    min=0.0,
    max=2.0,
    step=0.1,
    description='Brightness',
    disabled=True
)

contrast_slider = widgets.FloatSlider(
    value=1.0,
    min=0.0,
    max=2.0,
    step=0.1,
    description='Contrast',
    disabled=True
)

saturation_slider = widgets.FloatSlider(
    value=1.0,
    min=0.0,
    max=2.0,
    step=0.1,
    description='Saturation',
    disabled=True
)

# Dropdown for filters
filter_dropdown = widgets.Dropdown(
    options=['None', 'Grayscale', 'Sepia', 'Invert'],
    description='Filter',
    disabled=True
)

# Buttons for generating and modifying the image
generate_button = widgets.Button(description="Generate Image", button_style='primary')
modify_button = widgets.Button(description="Modify Image", button_style='info')
modify_button.disabled = True

image_box = widgets.Output()
download_box = widgets.Output()

# Bind button click events to functions
generate_button.on_click(on_generate_button_click)
modify_button.on_click(on_modify_button_click)

# Layout the UI
ui = widgets.VBox([
    widgets.HTML('<h2>Rich Text-to-Image Generator</h2>'),
    input_box,
    generate_button,
    widgets.HTML('<h3>Modify Image</h3>'),
    widgets.HBox([brightness_slider, contrast_slider, saturation_slider, filter_dropdown]),
    modify_button,
    image_box,
    download_box
])

# Display the UI
display.display(ui)
```
